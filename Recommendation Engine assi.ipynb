```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('C:\\Users\\SONY\\Downloads\\book (1).csv',encoding='latin-1')
movies_df.columns=['index', 'userId', 'booktitle', 'rating']
```

```python
movies_df=movies_df.iloc[:,1:]
movies_df.dtypes
```

```python
#number of unique users in the dataset
len(movies_df['userId'].unique())
```

```python
len(movies_df['booktitle'].unique())
```

```python
movies_df.set_index(['userId', 'booktitle', 'rating'], append=True)
```

```python
user_movies_df = movies_df.pivot_table(index='userId',
                                 columns='booktitle',
                                 values='rating', aggfunc='first').reset_index(drop=True)
```

```python
user_movies_df
```

```python
user_movies_df.index = movies_df.userId.unique()
```

```python
user_movies_df
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_movies_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
